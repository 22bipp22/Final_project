```python
# Import Dependencies
from bs4 import BeautifulSoup as bs
import pandas as pd
import requests
from sqlalchemy import create_engine
import os
```

```python
# Reading the poverty and income csv file and creating a dataframe
complete_csv_file ="complete.csv"
complete_df = pd.read_csv(complete_csv_file, encoding = "UTF-8")
complete_df
```

```python
#Reading the coordinates csv file and creating dataframe
coordinate_csv_file = "countycoordinates.csv"
coordinate_df = pd.read_csv(coordinate_csv_file, encoding = "UTF-8")
coordinate_df.rename(columns={'County Name': 'County'}, inplace=True)
coordinate_df
```

```python
# Reading the crime csv file and creating a dataframe
area_csv_file ="area.csv"
area_df = pd.read_csv(area_csv_file, encoding = "UTF-8")
area_df.rename(columns={'Name': 'County', "Total Area (mi2)": "Total_Area"}, inplace=True)
area_df = area_df.drop(columns=['Land Area (mi2)', 'PopulationDensity (mi2)'])
area_df
```

```python
merge_df = complete_df.merge(coordinate_df, on = "County")


merge_df
```

```python
merge_df = merge_df.merge(area_df, on = "County")
merge_df
```

```python
#convert to a json
merge_df.to_json('complete.json', index=True)
```
