```python
import pandas as pd
import numpy as np
from pprint import pprint
from sqlalchemy import create_engine
from keys import sqlkey
```

```python
#CREATE DATABASE CONNECTION
engine = create_engine('postgresql://postgres:'+sqlkey+'@localhost:5432/horse_races')
connection = engine.connect()
```

```python
data_df = pd.read_sql(sql='select * from raceruns',con=connection)
```

```python
data_df = data_df.dropna()
```

```python
from sklearn.model_selection import train_test_split

X = data_df.drop(columns=["won"])

y = data_df["won"]


X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, stratify = y)
```

```python
X_train.head(50)

X_train.to_csv("X_train.csv")
y_train.to_csv("y_train.csv")

X_test.to_csv("X_test.csv")
y_test.to_csv("y_test.csv")
```

```python
from sklearn import preprocessing 
scaler = preprocessing.MinMaxScaler()
minmax_df = scaler.fit(X_train) #something wrong here
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
horse_model = LogisticRegression(max_iter=10000)
horse_model
```

```python
horse_model.fit(X_train_scaled, y_train)
```

```python
print(f"Training Data Score: {horse_model.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {horse_model.score(X_test_scaled, y_test)}")
print(f"for one single line:{horse_model.predict(X_train_scaled[12].reshape(1,-1))}")
```
